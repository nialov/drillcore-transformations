```python
borehole_vector = vector_from_dip_and_dir(borehole_plunge, borehole_trend)
print(borehole_vector)
```

```python
gamma_vector = rotate_vector_about_vector(plane_vector, plane_normal, gamma)
print(gamma_vector)
```
